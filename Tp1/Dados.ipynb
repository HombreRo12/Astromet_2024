Son pseudoaleatorios, se generan a partir de una ecuación que tiene una IC conocida como semilla. Dada una semilla se me genera una serie de números pseudoaleatorios.
Si pasa ciertos test de aleatoriedad donde esencialmente no se encuentran patrones entre ellos, se puede considerar que son a efectos prácticos aleatorios.
Se usa esto para tener control en la generación. Para elegir la semilla usamos np.random.seed

Los generadores generan un número finito (muy grande) de datos y luego empiezan a repetirse, son cíclicos. Se puede cambiar la semilla para que no se repita la secuencia.

Son distribuciones homogeneas, se les puede aplicar transformaciones para convertirlas en Gaussianas, Poisson, Lorentz, etc.

```python
import numpy as np
import matplotlib.pyplot as plt

m = int(input("Ingrese el número de veces que desea lanzar los dados: "))
dados = np.random.randint(1, 7, size=(m, 100)) #En el otro podría haber puesto donde dice 100, 2, en lugar de definir 2 dados. Así se arma una matríz de 100 dados que se lanzan m veces.
suma_dados = np.sum(dados, axis=1) #axis=1 hace que sume en columnas (axis=0 lo haría en filas)

print(f"Número de lanzamientos: {m}")
print(f"Suma de los 100 dados en cada lanzamiento:\n{suma_dados}")

bins = np.arange(100, 601) - 0.5
plt.hist(suma_dados, density=True, bins=bins, edgecolor='black')

plt.title('Distribución de la suma al lanzar 100 dados')
plt.xlabel('Suma')
plt.ylabel('Frecuencia relativa')
plt.xticks(np.arange(100, 601, 50))
plt.grid(axis='y', alpha=0.75)

plt.show()


media = np.median(suma_dados)
desvest = np.std(suma_dados)

print(media)
print(desvest)
```
