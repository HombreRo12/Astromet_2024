```python
from sklearn.neural_network import MLPClassifier
from	sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, classification_report


features = ['U-G', 'G-R','i-z', 'Hmag','Jmag','umag','gmag','rmag','imag','zmag','Ymag','Kmag','W1mag','W2mag','u3mag','g3mag','r3mag','i3mag','Z3mag','Y3mag','J3mag','H3mag',
            'K3mag', 'J3/J_r', 'Y3/Y_r', 'Yhlr', 'Jhlr', 'Hhlr', 'Khlr',
            'H-K', 'H-J', 'J-K', 'Y-J', 'Y-K', 'J-H', 'K-H', 'K-J', 'Y-H', 'H-W1','J-W1','K-W1','Y-W1','H-W2','J-W2','K-W2',
            'Y-W2','H-W1','J-W1','K-W1','Y-W1','H-W2','J-W2','K-W2','Y-W2','H-W1','J-W1','K-W1','Y-W1','H-W2','J-W2','K-W2',
            'Y-W2','H-W1','J-W1'] 



df_train, df_test = train_test_split(train, test_size=0.3, random_state=42)
X_train = df_train[features]
Y_train	= df_train['Hclass']
X_test	= df_test[features]
Y_test	= df_test['Hclass']

clf = MLPClassifier(hidden_layer_sizes=(100,100,100), max_iter=500, alpha=0.0001, activation='tanh',  random_state=21,learning_rate='adaptive')
```

```python
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred, average='weighted'))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X_train, Y_train, cv=5, scoring='f1_weighted', n_jobs=-1)
print(scores)
```
