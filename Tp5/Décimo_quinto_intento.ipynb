```python
import numpy as np
import	matplotlib.pyplot as plt
import pandas as pd
from sklearn.impute import KNNImputer


train = pd.read_csv('train.csv')
train.set_index('ID',	inplace=True)

#Cambio a Nans
problema = ['Yhlr', 'Jhlr', 'Hhlr', 'Khlr']

train[problema] = train[problema].applymap(lambda x: np.nan if x <-30 or x>30 else x)
imputer = KNNImputer(missing_values=np.nan, n_neighbors=5, weights='distance',)

train_estrella = train[train['Hclass'] == 0]
train_galaxia = train[train['Hclass'] == 1]
train_qso = train[train['Hclass'] == 2]

train_estrella[problema] = imputer.fit_transform(train_estrella[problema])
train_galaxia[problema] = imputer.fit_transform(train_galaxia[problema])
train_qso[problema] = imputer.fit_transform(train_qso[problema])

train = pd.concat([train_estrella, train_galaxia, train_qso])
train = train.sample(frac=1, random_state=4)





#Magnitudes, para armar colores
H = train.loc[:,'Hmag']
J	= train.loc[:,'Jmag']
U =	train.loc[:,'umag']
G =	train.loc[:,'gmag']
r = train.loc[:,'rmag']
i = train.loc[:,'imag']
z = train.loc[:,'zmag']
Y = train.loc[:,'Ymag']
K = train.loc[:,'Kmag']
W1 =	train.loc[:,'W1mag']
W2 =	train.loc[:,'W2mag']

#Radios donde se encierra la mitad de la luz

Y_r =	train.loc[:,'Yhlr']
J_r =	train.loc[:,'Jhlr']
H_r =	train.loc[:,'Hhlr']
K_r =	train.loc[:,'Khlr']

#Magnitud en apertura fija de 3''

U3 =	train.loc[:,'u3mag']
G3 =	train.loc[:,'g3mag']
r3 =	train.loc[:,'r3mag']
i3 =	train.loc[:,'i3mag']
Z3	=	train.loc[:,'Z3mag']
Y3	=	train.loc[:,'Y3mag']
J3	=	train.loc[:,'J3mag']
H3	=	train.loc[:,'H3mag']
K3	=	train.loc[:,'K3mag']

#Coordenadas

ra = train.loc[:,'RAdeg']
dec = train.loc[:,'DEdeg']

#Clases para entrenar

clase	= train.loc[:,'Hclass'] #0: Estrella, 1: Galaxia, 2: QSO






# Agregar columnas U-G, G-R, i-z, J3/J_r y Y3/Y_r
train['U-G'] = U - G
train['G-R'] = G - r
train['i-z'] = i - z
train['J3/J_r'] = J3 / J_r
train['Y3/Y_r'] = Y3 / Y_r
train['W2-W1'] = W2 - W1
train['H-K'] = H - K
train['H-J'] = H - J
train['J-K'] = J - K
train['Y-J'] = Y - J
train['Y-K'] = Y - K
train['J-H'] = J - H
train['K-H'] = K - H
train['K-J'] = K - J
train['Y-H'] = Y - H
train['H-W1'] = H - W1
train['J-W1'] = J - W1
train['K-W1'] = K - W1
train['Y-W1'] = Y - W1
train['H-W2'] = H - W2
train['J-W2'] = J - W2
```

```python
train.describe()
```

```python
from sklearn.ensemble import RandomForestClassifier
from	sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score, confusion_matrix, classification_report


features = ['U-G', 'G-R','i-z', 'Hmag','Jmag','umag','gmag','rmag','imag','zmag','Ymag','Kmag','W1mag','W2mag','u3mag','g3mag','r3mag','i3mag','Z3mag','Y3mag','J3mag','H3mag',
            'K3mag', 'J3/J_r', 'Y3/Y_r', 'Yhlr', 'Jhlr', 'Hhlr', 'Khlr',
            'H-K', 'H-J', 'J-K', 'Y-J', 'Y-K', 'J-H', 'K-H', 'K-J', 'Y-H', 'H-W1', 'J-W1', 'K-W1',] 



df_train, df_test = train_test_split(train, test_size=0.3, random_state=42)
X_train = df_train[features]
Y_train	= df_train['Hclass']
X_test	= df_test[features]
Y_test	= df_test['Hclass']

clf = RandomForestClassifier(n_estimators=150, random_state=42,min_samples_split=15, min_samples_leaf=3,
                             max_depth=15, criterion='entropy',bootstrap=True, n_jobs=-1)
```

```python
clf.fit(X_train, Y_train)
Y_pred = clf.predict(X_test)
print(confusion_matrix(Y_test, Y_pred))
print(classification_report(Y_test, Y_pred))
print(f1_score(Y_test, Y_pred, average='weighted'))
```

```python
pd.DataFrame(clf.feature_importances_*100, index=features)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X_train, Y_train, cv=5, scoring='f1_weighted', n_jobs=-1)
print(scores)
```
